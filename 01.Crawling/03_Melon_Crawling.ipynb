```python
#크롤링 하는게 봇이 아닌 사람이므로 접근 허용할 수 있도록 하는 작업
headers= {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
url = 'https://www.melon.com/chart/index.htm'
req = requests.get(url, headers=headers)
req.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.text, 'html.parser')
```

### 01.원하는 데이터 찾기 (tr형태)

```python
trs = soup.find_all('tr')
len(trs)
```

```python
#1번째 방식 trs = soup.selector('.service_list_song.type02.d_song_list > table > tbdoy > tr')
trs = soup.select('#frm > div > table > tbody > tr')
len(trs)
```

### 2. 하나를 선택해서 원하는 정보 추출

```python
tr = trs[0]
```

```python
#순위
rank=int(tr.select_one('.rank').get_text().strip())
rank
```

```python
#제목
title = tr.select_one('.ellipsis.rank01').get_text().strip()
title
```

```python
#가수
artist = tr.select_one('.ellipsis.rank02 > a').get_text().strip()
#artist = tr.select_one('.ellipsis.rank02').find('a').get_text().strip()
artist
```

```python
# 앨범
album = tr.select_one('.ellipsis.rank03').get_text().strip()
album
```

```python
#좋아요 - > selenium을 통해서 가져와야함(동적스크래핑)
Like = tr.select_one('.')
Like
```

### 3. 반복문을 통한 결과물 완성(데이터 가져오기)

```python
lines= []
for tr in trs:
  rank=int(tr.select_one('.rank').get_text().strip())
  title = tr.select_one('.ellipsis.rank01').get_text().strip()
  artist = tr.select_one('.ellipsis.rank02 > a').get_text().strip()
  album = tr.select_one('.ellipsis.rank03').get_text().strip()
  lines.append([rank,title,artist,album]) #2차원 리스트로 형태 만들기
```

### 4.Dataframe화 하기

```python
import pandas as pd
df= pd.DataFrame(lines, columns=['순위','제목','가수','앨범'])
```
